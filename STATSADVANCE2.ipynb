Q1- DEFINE THE Z-STATISTICS AND EXPLAIN ITS RELATIONSHIP TO THE STANDARD NORMAL DISTRIBUTION . HOW IS THE Z STATISTICS USED IN HYPOTHESIS TESTING?

ANSWER-

* Z-STATISTIC: THE Z-STATISTIC, ALSO KNOWN AS THE Z-SCORE, IS A MEASURE THAT DESCRIBES THE POSITION OF A RAW SCORE IN TERMS OF ITS DISTANCE FROM THE MEAN, EXPRESSED IN STANDARD DEVIATION UNITS. IT IS USED TO DETERMINE HOW MANY STANDARD DEVIATIONS AN ELEMENT IS FROM THE MEAN

* RELATIONSHIP TO THE STANDARD NORMAL DISTRIBUTION: THE Z-STATISTIC STANDARDIZES A RAW SCORE TO THE STANDARD NORMAL DISTRIBUTION, WHICH HAS A MEAN OF 0 AND A STANDARD DEVIATION OF 1. THIS ALLOWS FOR COMPARISON OF SCORES FROM DIFFERENT DISTRIBUTIONS AND IS USEFUL IN CALCULATING PROBABILITIES.

**THE Z-STATISTIC IS USED IN HYPOTHESIS TESTING TO DETERMINE WHETHER THERE IS ENOUGH EVIDENCE TO REJECT A NULL HYPOTHESIS**

Q3- COMPARE AND CONTRAST THE BIONOMIAL AND BERNOULLI DISTRIBUTIONS.

ANSWER - BIONOMIAL DISTRIBUTION VS BERNOULLI DISTRIBUTION -

BIONOMIAL DISTRIBUTION -

* TRIALS : MULTIPLE (n)
* OUTCOME : NUMBER OF SUCCESSES
* PARAMETER : n,p
* RANGE : 0,1,2,....,N
* MEAN : np
* VARIANCE : np(1-p)

---

BERNOULLIS DISTRIBUTION -

* TRIALS : SINGLE
* OUTCOME : SUCCESS OR FAILURE
* PARAMETER : p
* RANGE : 0 or 1
* MEAN : p
* VARIANCE : p(1-p)

MAIN DIFFERENCE - BIONOMIAL MEANS MULTIPLE BERNOULLI TRIALS COMBINED WHERE BERNOULLI REFERS TO SINGLE TRIAL.

Q5- WHAT ARE THE KEY PROPERTIES OF THE POISSON DISTRIBUTION, AND WHEN IS IT APPROPRIATE TO USE THIS DISTRIBUTION ?

1. DISCRETE DISTRIBUTION: DEALS WITH THE NUMBER OF EVENTS IN A FIXED INTERVAL OF TIME OR SPACE.

4. MEMORYLESS: THE OCCURRENCE OF AN EVENT IS INDEPENDENT OF THE OCCURRENCE OF PREVIOUS EVENTS.

5. USES NATURAL NUMBERS: VALUES RANGE FROM 0 TO ∞.

* WHEN TO USE THE POISSON DISTRIBUTION :

1) MODELING THE NUMBER OF TIMES AN EVENT OCCURS IN A FIXED INTERVAL, SUCH AS THE NUMBER OF CALLS RECEIVED BY A CALL CENTER PER HOUR.

2) EVENTS MUST BE INDEPENDENT AND OCCUR WITH A CONSTANT AVERAGE RATE.

Q6- DEFINE THE TERMS "PROBABILITY DISTRIBUTION" AND PROBABILITY DENSITY FUNCTION " . HOW DOES A PDF DIFFER FROM A PROBABILITY MASS FUNCTION?

ANSWER- PROBABILITY DISTRIBUTION AND PDF
PROBABILITY DISTRIBUTION:

A FUNCTION THAT GIVES THE PROBABLITIES OF OCCURRENCE OF DIFFERENT POSSIBLE OUTCOMES IN AN EXPERIMENT. BASICALLY, IT TELLS YOU HOW LIKELY DIFFERENT OUTCOMES ARE.

* PROBABILITY DENSITY FUNCTION :

A FUNCTION THAT DESCRIBES THE LIKELIHOOD OF A CONTINUOUS RANDOM VARIABLE TO TAKE ON A PARTICULAR VALUE. THE AREA UNDER THE PDF CURVE FOR A GIVEN INTERVAL REPRESENTS THE PROBABILTY THAT THE VARIABLE FALLS WITHIN THAT INTERVAL. NOTE, THE PDF ITSELF IS NOT A PROBABILITY, BUT THE AREA UNDER THE CURVE IS.

PDF VS. PROBABILITY MASS FUNCTION
* PROBABILITY MASS FUNCTION:

USED FOR DISCRETE RANDOM VARIABLES. IT GIVES THE PROBABILITY THAT A DISCRETE RANDOM VARIABLE IS EXACTLY EQUAL TO SOME VALUE. FOR INSTANCE, THE PMF FOR A FAIR DIE ROLL GIVES YOU THE PROBABLITY OF ROLLING A 1, 2, 3, ETC.

DIFFERENCES:-

CONTINUOUS VS. DISCRETE: PDF IS USED FOR CONTINUOUS VARIABLES, PMF IS FOR DISCRETE VARIABLES.

VALUES GIVEN: PDF GIVES LIKELIHOOD , PMF GIVES ACTUAL PROBABILITIES.

TOTAL AREA/SUM: THE INTEGRAL OF A PDF OVER ITS ENTIRE RANGE IS 1, WHILE THE SUM OF ALL PROBABILITIES GIVEN BY A PMF IS 1.

Q8- COMPARE Z-SCORE AND T-SCORES. WHEN SHOULD YOU USE A Z-SCORE, AND WHEN SHOULD A T-SCORE BE APPLIED INSTEAD?

ANSWER - Z-SCORE VS. T-SCORE
Z-SCORE:

DEFINITION: MEASURES THE NUMBER OF STANDARD DEVIATIONS AN ELEMENT IS FROM THE MEAN.
USAGE: WHEN THE SAMPLE SIZE IS LARGE (TYPICALLY 𝑛>30) AND THE POPULATION STANDARD DEVIATION IS KNOWN.

DISTRIBUTION: FOLLOWS THE STANDARD NORMAL DISTRIBUTION (MEAN = 0, STANDARD DEVIATION = 1).

T-SCORE:

DEFINITION: MEASURES THE NUMBER OF STANDARD DEVIATIONS AN ELEMENT IS FROM THE MEAN IN A SAMPLE.

USAGE: WHEN THE SAMPLE SIZE IS SMALL (TYPICALLY N < 30) OR WHEN THE POPULATION STANDARD DEVIATION IS UNKNOWN.

DISTRIBUTION: FOLLOWS THE T-DISTRIBUTION, WHICH HAS HEAVIER TAILS THAN THE NORMAL DISTRIBUTION .

WHEN SHOULD WE USE -

Z-SCORE: USE WHEN WE HAVE A LARGE SAMPLE SIZE AND KNOW THE POPULATION STANDARD DEVIATION.

T-SCORE: USE WHEN WE HAVE A SMALL SAMPLE SIZE OR DO NOT KNOW THE POPULATION STANDARD DEVIATION.

Q9-GIVEN A SAMPLE MEAN OF 105 , A POPULATION MEAN OF 100, A STANDARD DEVIATION OF 15 , AND A SAMPLE SIZE OF 25, CALCULATE THE Z-SCORE AND P-VALUE . BASED ON A SIGNIFICANCE LEVEL OF 0.05, DO YOU REJECT OR FAIL TO REJECT THE NULL HYPOTHESIS?
TASK: WRITE PYTHON CODE TO CALCULATE THE Z-SCORE AND P-VALUE FOR THE GIVEN DATA.
OBJECTIVE: APPLY THE FORMULA FOR THE Z-SCORE AND  INTERPRET THE P-VALUE FOR HYPOTHESIS TESTING.

ANSWER -

```python
import scipy.stats as stats

# GIVEN DATA
sample_mean = 105
population_mean = 100
std_deviation = 15
sample_size = 25

# CALCULATING THE Z-SCORE
z_score = (sample_mean - population_mean) / (std_deviation / (sample_size ** 0.5))

# CALCULATING THE P-VALUE
p_value = 1 - stats.norm.cdf(z_score)

# PRINT THE RESULTS
print("Z-Score:", z_score)
print("P-Value:", p_value)

# INTERPRETATION
anurag = 0.05
if p_value < anurag:
    print("REJECT THE NULL HYPOTHESIS")
else:
    print("FAIL TO REJECT THE NULL HYPOTHESIS")
```

Q10- SIMULATE A BIONOMIAL DISTRIBUTION WITH 10 TRIALS AND A PROBABILITY 0F SUCCESS OF 0.6 USING PYTHON .
GENERATE 1000 SAMPLES AND PLOT THE DISTRIBUTION . WHAT IS THE EXPECTED MEAN AND VARIANCE?
TASK: USE PYTHON CODE TO GENERATE THE DATA , PLOT THE DISTRIBUTION , AND CALCULATE THE MEAN AND VARIANCE.
OBJECTIVES: UNDERSTAND THE PROPERTIES OF A BIONOMIAL DISTRIBUTION AND VERIFY THEM THROUGH SIMULATION.

ANSWER -

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.6
num_samples = 1000

samples = np.random.binomial(n, p, num_samples)

mean = np.mean(samples)
variance = np.var(samples)

expected_mean = n * p
expected_variance = n * p * (1 - p)

plt.hist(samples, bins=range(n+2), density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.6)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

print("Sample Mean:", mean)
print("Sample Variance:", variance)
print("Expected Mean:", expected_mean)
print("Expected Variance:", expected_variance)
```
