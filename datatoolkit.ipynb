# Q3 -  EXPLAIN THE FOLLOWING TERMS :
*   THE DIFFERENCE IN np.array, np.asarray and np.asanarray.
*   THE DIFFERENCE BETWEEN DEEP COPY AND SHALLOW COPY .


# ANSWERS - THE DIFFERENCE IN np.array, np.asarray and np.asanarray.

ANSWER-
1) np.array - THE np.array FUNCTION IN NumPY CREATES A NEW N-DIMENSIONAL ARRAY FROM INPUT DATA SUCH AS A LIST, TUPLE , OR ANOTHER SEQUENCE TYPE .BY DEFAULT IT MAKES A COPY OF THE INPUT DATA , ENSURING THAT MODIFICATIONS TO THE NEW ARRAY WON'T AFFECT THE ORIGINAL DATA .

```python
import numpy as np
my_list = [1,2,3,4,5]
arr = np.array(my_list)
print(arr)
```

2) np.asarray() -

*  IT ALSO CONVERT INPUT DATA TO A NumPy array .UNLIKE np.array(). IT DOES NOT ALWAYS MAKE A COPY . IF THE INPUT IS ALREADY AN ndarray , IT RETURNS A VIEW OF THE ORIGINAL DATA WITHOUT COPYING.

```python
# FOR EXAMPLE -
import numpy as np
my_list = [1,2,3,4,5]
arr = np.asarray(my_list)
print(arr)
```

3)np.asanarray()-
IT IS SIMILAR TO np.asarray(). HOWEVER , IT HAS ONE KEY DIFFERENCE : IF THE INPUT IS SUBCLASS OF ndarray , IT WILL RETURN AN ARRAY OF THAT SUBCLASS OTHERWISE , IT BEHAVES LIKE np.addarray() AND CREATES A VIEW OR A NEW ARRAY .

```python
# FOR EXAMPLE -
my_matrix = np.matrix([[1, 2], [3, 4]])
arr_subclass = np.asanyarray(my_matrix)
print(arr_subclass)
```

#Q14 - CREATE A TIME - SERIES DATASET IN A PANDAS DATAFRAME WITH COLUMNS:'DATE', ' TEMPERATURE' , 'HUMIDITY' AND PERFORM THE FOLLOWING TASKS USING MATPLOTLIB:

A) PLOT THE 'TEMPERATURE' AND 'HUMIDITY' ON THE SAME PLOT WITH DIFFERENT y-axis ( left y-axis for 'Temperature ' and right y-axis for 'Humidity')

B)Label the x-axis AS 'Date'.

C)Set the title of the plot as 'Temperature and Humidity Over Time'.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Creating a DataFrame with 'DATE', 'TEMPERATURE', and 'HUMIDITY'
data = {'DATE': date_range,'TEMPERATURE': np.random.uniform(low=15, high=35, size=100),'HUMIDITY': np.random.uniform(low=40, high=90, size=100)}
df = pd.DataFrame(data)
# Plotting the 'TEMPERATURE' and 'HUMIDITY' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plotting temperature on the left y-axis
ax1.plot(df['DATE'], df['TEMPERATURE'], color='red', label='Temperature'),ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax2 = ax1.twinx()
ax2.plot(df['DATE'], df['HUMIDITY'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
plt.title('Temperature and Humidity Over Time')
plt.legend()
plt.show()

```

#Q15 - CREATE A NUMPY ARRAY DATA CONTAINING 1000 SAMPLES FROM A NORMAL DISTRIBUTION . PERFORM THE FOLLOWING TASKS USING MATPLOTLIB:

A) PLOT A HISTOGRAM OF THE DATA WITH 30 BINS.

B)OVERLAY A LINE PLOT REPRESENTING THE NORMAL DISTRIBUTION'S PROBABILITY DENSITY FUNCTION (pdf).

c) LABEL THE X-AXIS AS 'VALUE' AND Y-AXIS AS 'FREQUENCY/PROBABILITY'.

D) SET THE TITLE OF THE PLOT AS 'HISTOGRAM WITH PDF OVERLAY'.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')
plt.xlabel('VALUE')
plt.ylabel('FREQUENCY/PROBABILITY')
plt.legend()
plt.show()
```

# Q16 - SET THE TITLE OF THE PLOT AS ' HISTOGRAM WITH PDF OVERLAY ' .

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, loc=0, scale=1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')
plt.title('HISTOGRAM WITH PDF OVERLAY')
plt.xlabel('VALUE')
plt.ylabel('FREQUENCY/PROBABILITY')
plt.legend()
plt.show()
```

# Q17 - CREATE A SEABORN SCATTER PLOT OF TWO RANDOM ARRAYS , COLOR POINTS BASED ON THEIR POSITION RELATIVE TO THE ORIGIN(quadrant) , ADD A LEGEND , LABEL THE AXES , AND SET THE TITLE AS ' QUADRANT-WISE SCATTER PLOT'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
quadrants = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        quadrants.append('Q1')
    elif x[i] < 0 and y[i] >= 0:
        quadrants.append('Q2')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Q3')
    else:
        quadrants.append('Q4')
data = {'x': x, 'y': y, 'quadrant': quadrants}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='Set1', s=100)
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.show()
```

# Q18- WITH BOKEH , PLOT A LINE CHART OF A SINE WAVE FUNCTION , ADD GRID LINES , LABEL THE AXES , AND SET THE TITLE AS 'SINE WAVE FUNCTION'.

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)
output_notebook()
p = figure(title="SINE WAVE FUNCTION", x_axis_label='X-AXIS', y_axis_label='Y-AXIS')
p.line(x, y, legend_label="Sine Wave", line_width=2)
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
show(p)
```

# Q19 - USING BOKEH , GENERATE A BAR CHART OF RANDOMLY GENERATED CATEGORICAL DATA, COLOR BARS USED ON THEIR VALUES, ADD HOVER TOOLTIPS TO DISPLAY EXACT VALUES , LABEL THE AXES , AND SET THE TITLE AS 'RANDOM CATEGORICAL BAR CHART'.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
output_notebook()
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))
data = pd.DataFrame({'categories': categories, 'values': values})
source = ColumnDataSource(data)
p = figure(x_range=categories, title="RANDOM CATEGORICAL BAR CHART", toolbar_location=None, tools="")
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"
show(p)
```

# Q21 - USING PLOTLY , CREATE AN INTERACTIVE PIE CHART OF RANDOMLY GENERATED DATA , ADD LABELS AND PERCENTAGES , SET THE TITLE AS ' INTERACTIVE PIE CHART'.

```python
import plotly.express as px
import numpy as np
import pandas as pd
np.random.seed(0)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(10, 100, size=len(categories))
df = pd.DataFrame({'Category': categories, 'Value': values})
fig = px.pie(df, values='Value', names='Category', title='INTERACTIVE PIE CHART',
labels={'Category': 'Category', 'Value': 'Value'},
hover_data={'Value': ':.2f'})
fig.update_traces(textinfo='label+percent', insidetextorientation='radial')
fig.show()
```
